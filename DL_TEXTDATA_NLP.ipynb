```python
from datasets import DatasetDict

# Define the split ratio (e.g., 20% of validation becomes test)
test_split = dataset["validation"].train_test_split(test_size=0.2, seed=42)

# Create new validation and test sets
new_validation_set = test_split["train"]
testDataset = test_split["test"]

# Update dataset dictionary
dataset = DatasetDict({
    "train": dataset["train"],
    "validation": new_validation_set,
    "test": testDataset
})

# Print dataset splits
print(dataset)
```
